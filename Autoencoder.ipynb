```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Student marks
X = np.array([
    [85, 90, 88, 80, 92],
    [70, 75, 72, 78, 80],
    [95, 92, 96, 90, 94],
    [60, 65, 62, 70, 68]
])

# Normalize
X = X / 100

# Autoencoder Model
model = Sequential([
    Dense(2, activation='relu', input_shape=(5,)),   # Encoder
    Dense(5, activation='sigmoid')                   # Decoder
])

# Compile
model.compile(optimizer='adam', loss='mse')

# Train
model.fit(X, X, epochs=100, verbose=0)

# Predict
output = model.predict(X)

# Display Results
print("Original Marks:")
print(X[0] * 100)

print("\nReconstructed Marks:")
print(output[0] * 100)
```
